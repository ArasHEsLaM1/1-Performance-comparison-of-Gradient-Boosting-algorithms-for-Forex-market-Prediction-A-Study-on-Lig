```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
```

```python

# Load the dataset
df = pd.read_csv("customer_data.csv")

# Select the features for clustering
X = df.iloc[:, 2:].values
```

```python

# Perform K-means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Perform Agglomerative clustering
agglomerative = AgglomerativeClustering(n_clusters=5)
agglomerative_labels = agglomerative.fit_predict(X)

# Perform DBSCAN clustering
dbscan = DBSCAN(eps=3, min_samples=4)
dbscan_labels = dbscan.fit_predict(X)
```

```python


# Add the clustering labels to the dataset
df['KMeans_Labels'] = kmeans_labels
df['Agglomerative_Labels'] = agglomerative_labels
df['DBSCAN_Labels'] = dbscan_labels
```

```python



# Print the clustering labels for each customer
print("K-means Clustering Labels:")
print(df[['CustomerID', 'KMeans_Labels']].head())
```

```python

print("\nAgglomerative Clustering Labels:")
print(df[['CustomerID', 'Agglomerative_Labels']].head())
```

```python

print("\nDBSCAN Clustering Labels:")
print(df[['CustomerID', 'DBSCAN_Labels']].head())
```

```python

# Visualize the clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=kmeans_labels, palette='viridis')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('K-means Clustering')
plt.legend(title='K-means Clustering Labels', loc='upper right')
plt.show()

plt.figure(figsize=(10, 8))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=agglomerative_labels, palette='viridis')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('Agglomerative Clustering')
plt.legend(title='Agglomerative Clustering Labels', loc='upper right')
plt.show()

plt.figure(figsize=(10, 8))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=dbscan_labels, palette='viridis')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('DBSCAN Clustering')
plt.legend(title='DBSCAN Clustering Labels', loc='upper right')
plt.show()
```
